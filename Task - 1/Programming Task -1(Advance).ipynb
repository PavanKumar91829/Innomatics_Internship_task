### 1. Given an array `nums`. We define a running sum of an array as:
`runningSum[i] = sum(nums[0] ... nums[i])
`

**Exple 1**
- **Input:** `nums = [1, 2, 3, 4]`
- **Output:** `[1, 3, 6, 10]`
- **Explanation:** Running sum is computed as:
  `[1, 1+2, 1+2+3, +2+3+4]`

### 2. Shuffle the Array

Given the array `nums` consisting of `2n` elements in the form  
`[x1, x2, ..., xn, y1, y2, ..., yn]`.

Return the array in the form:  
`[x1, y1, x2, y2, ..., xn, yn]`.

**Example**
- **Input:** `nums = [2,5,1,3,4,7]`, `n = 3`
- **Output:** `[2,3,5,4,1,7]`
- **Explanation:**  
  Since `x1=2`, `x2=5`, `x3=1`, `y1=3`, `y2=4`, `y3=7`,  
  the result is `[2,3,5,4,1,7]`.


```python
nums = [2,5,1,3,4,7]
n = 3
ans=[]
for i in range(n):
    ans.append(nums[i])
    ans.append(nums[i+n])
print(ans)        
```

## 3. Kids With the Greatest Number of Candies

There are `n` kids with candies.  
You are given an integer array `candies`, where `candies[i]` is the number of candies the *i*-th kid has,  
and an integer `extraCandies`, representing the number of extra candies available.

Return a boolean array `result` of length `n`, where `result[i]` is `true` if, after giving the *i*-th kid **all** the `extraCandies`,  
they would have the greatest number of candies among all kids; otherwise, `false`.

Note: Multiple kids may tie for the greatest number of candies.

### Example 1
- **Input:** `candies = [2,3,5,1,3]`, `extraCandies = 3`
- **Output:** `[true, true, true, false, true]`


```python
candies = [2,3,5,1,3]
extraCandies = 3
ans = []
for c in candies:
    if c+extraCandies>=max(candies):
        ans.append(True)
    else:
        ans.append(False)
print(ans)
```
